```python
import tensorflow.keras as k
from tensorflow.keras import layers, activations, models, preprocessing, utils

encoder_inputs = k.layers.Input(shape=(None,))
encoder_embedding = k.layers.Embedding(310, 256,mask_zero=True )(encoder_inputs)
encoder_outputs, state_h, state_c = k.layers.LSTM(128, return_state=True)(encoder_embedding)
encoder_states = [state_h, state_c]


decoder_inputs = k.layers.Input(shape=(None,))
decoder_embedding = k.layers.Embedding(310, 256,mask_zero=True )(decoder_inputs)
decoder_lstm = k.layers.LSTM(128, return_state=True, return_sequences=True)
decoder_outputs, _ , _ , = decoder_lstm(decoder_embedding, initial_state=encoder_states)
decoder_dense = k.layers.Dense(307 ,activation=k.activations.softmax)
output = decoder_dense(decoder_outputs)



model = k.models.Model([encoder_inputs,decoder_inputs],output)
model.compile(optimizer=k.optimizers.RMSprop(), loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
#remove stop words
def without_stop_words(sentence) :
    for i in range(len(sentence)):
        tokens = sentence.split()  
        filtered_tokens = [token for token in tokens if token not in stop_words] 
        new_text = ' '.join(filtered_tokens) 
        sentence = new_text 
    
    return sentence
```

```python
#lemmatization
def to_stem(sentence) :
    from farasa.stemmer import FarasaStemmer
    stemmer = FarasaStemmer()

    S1 = []
    S = stemmer.stem(sentence)
    stemmed_sentence = S.split()

    return stemmed_sentence
```
